```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.ndimage import label
import scipy.io as sio
from collections import Counter
from scipy.stats import zscore, wilcoxon
```

## data curation

```python
# Remove values from e1times that are too close (less than 60 apart, which is 6s as the smoothed data is in 1/10s)
def remove_close_values(arr, min_distance=60):
    if len(arr) == 0:
        return arr
    
    # Sort the array to make it easier to find close values
    sorted_indices = np.argsort(arr)
    sorted_arr = arr[sorted_indices]
    
    # Find which values to keep
    keep_mask = np.ones(len(sorted_arr), dtype=bool)
    
    i = 0
    while i < len(sorted_arr) - 1:
        # Check if current and next values are too close
        if sorted_arr[i+1] - sorted_arr[i] < min_distance:
            # Mark both for removal
            keep_mask[i] = False
            keep_mask[i+1] = False
            
            # Check if there are more consecutive close values
            j = i + 1
            while j < len(sorted_arr) - 1 and sorted_arr[j+1] - sorted_arr[j] < min_distance:
                keep_mask[j+1] = False
                j += 1
            
            # Skip to the next unchecked value
            i = j + 1
        else:
            i += 1
    
    # Return only the values we want to keep, in original order
    original_keep_mask = np.zeros(len(arr), dtype=bool)
    original_keep_mask[sorted_indices[keep_mask]] = True
    
    return arr[original_keep_mask]
```

```python
def convert_boundary_to_seconds(boundary_mark):
    # Get the length of the original array
    n = len(boundary_mark)
    
    # Calculate the length of the new array
    new_len = n // 10
    
    # Initialize the new array with zeros
    boundary_s = np.zeros(new_len)
    
    # Process each element in boundary_mark
    for index in range(n):
        if boundary_mark[index] != 0:  # non-zero value
            value = boundary_mark[index]
            remainder = index % 10
            
            if remainder < 5:
                target_index = index // 10
            else:
                target_index = index // 10 + 1
            
            # Make sure target_index is within bounds
            if target_index < new_len:
                boundary_s[target_index] = value
    return boundary_s
```

```python
def plot_boundary(smd1,boundary_mark):
    plt.figure(figsize=(10, 4))
    plt.plot(smd1)
    plt.scatter(np.where(boundary_mark==3)[0], smd1[np.where(boundary_mark==3)[0]], 
          color='purple', label='High Agreement')
    plt.scatter(np.where(boundary_mark==2)[0], smd1[np.where(boundary_mark==2)[0]], 
          color='green', label='Moderate Agreement')
    plt.scatter(np.where(boundary_mark==1)[0], smd1[np.where(boundary_mark==1)[0]], 
          color='orange', label='Weak Agreement')

    plt.ylabel('Boundary Score')
    plt.title(f'movie{i}')
    plt.xlabel('Time')
    plt.legend()
    plt.show()
```

```python
folder='../../Raw Data/event_boundary/smoothed/'
filename="filmfest_retrospective_boundaries_peaks_w10_data_including_onsets_titles_for_Flory.mat"
data=sio.loadmat(folder+filename,simplify_cells=True)

time_ls={'01':41,'02':391,'03':529,'04':999,'05':1134,'06':41,'07':370,'08':636,'09':783,'10':948}
#movie start point relative to movie start point in description/prediction dataset
consensus_ls={'01':-1,'02':0,'03':-1,'04':0,'05':0,'06':-1,'07':0,'08':0,'09':0,'10':0}
nbin=10
```

```python
#identify cutoffs for low, medium, and high agreetment for eventboundaries

smd1_total = np.array([])
movie_ls=['01','02','03','04','05','06','07','08','09','10']

for i in movie_ls:
    smd1=data['ebounds']['movie'+i]['dsmoothed']
    smd1_total=np.concatenate([smd1_total,smd1])
minimum_cutoff= np.percentile(smd1_total, 65)    
extract_events1, num_labels = label(smd1_total > minimum_cutoff)

# Initialize array for event times
if np.max(extract_events1) > 0:
    e1times = np.zeros(np.max(extract_events1))
else:
    e1times = np.array([])
    
# Find peak time in each event
for ee in range(1, np.max(extract_events1) + 1):
    zvec = np.zeros(smd1_total.shape)
    zvec[extract_events1 == ee] = smd1_total[extract_events1 == ee]
    e1times[ee-1] = np.argmax(zvec)
peak_values= smd1_total[list(e1times.astype(int))]
high_cutoff=np.percentile(peak_values,2/3*100)
medium_cutoff=np.percentile(peak_values,1/3*100)
```

```python
movie_ls=['01','02','03','04','05','06','07','08','09','10']
count= 0
for i in movie_ls:
    #identify cutoffs for low, medium, and high agreetment for eventboundaries
    smd1=data['ebounds']['movie'+i]['dsmoothed']
    boundary_mark=np.zeros(smd1.shape[0])

    # Main processing with 65th percentile
    extract_events1, num_labels = label(smd1 > minimum_cutoff)
    
    # Initialize array for event times
    if np.max(extract_events1) > 0:
        e1times = np.zeros(np.max(extract_events1))
    else:
        e1times = np.array([])
        
    # Find peak time in each event
    for ee in range(1, np.max(extract_events1) + 1):
        zvec = np.zeros(smd1.shape)
        zvec[extract_events1 == ee] = smd1[extract_events1 == ee]
        e1times[ee-1] = np.argmax(zvec)
    e1times_remove_overlap = remove_close_values(e1times) #remove close points
    #e1times = e1times[e1times>50] #remove title time

    
    # Convert e1times to integer indices (in case they're floats)
    peak_indices = e1times.astype(int)
    
    # For each peak time, assign values based on cutoffs
    for idx in peak_indices:
        if idx < len(smd1):  # Make sure index is within bounds
            value_at_peak = smd1[idx]            
            if value_at_peak > high_cutoff:
                boundary_mark[idx] = 3
            elif value_at_peak > medium_cutoff:
                boundary_mark[idx] = 2
            else:
                boundary_mark[idx] = 1



    #binning data into secs timeresolution
    boundary_s=convert_boundary_to_seconds(boundary_mark)

    # Create a mask for locations NOT in e1times_remove_overlap
    mask = ~np.isin(np.arange(len(boundary_mark)), e1times_remove_overlap)
    # Set those locations to 0
    boundary_mark_filter =boundary_mark.copy()
    boundary_mark_filter[mask] = 0
    boundary_s_filtered=convert_boundary_to_seconds(boundary_mark_filter)
    smoothed_score=convert_boundary_to_seconds(smd1)

    counts=Counter(boundary_s_filtered)
    count+=counts[3]+counts[2]+counts[1]

    high_boundary = boundary_s==3
    filmfest_offset=range(time_ls[i],len(boundary_s)+time_ls[i]) # timing in description/prediction scale
    tr_offset=[round(i/1.5) for i in list(filmfest_offset)] # timing in neuro data scale
    consensus_offset = range(consensus_ls[i],len(high_boundary)+consensus_ls[i])
    # boundary_score: all boundaries identified, scored by high/moderate/week agreement on boundary
    # boundary_s_filtered: boundary remained after filter by previous work's standard (if 2 boundary are less than 6s apart, remove)
    # strong_boundary: boundary_score==3
    boundary_df=pd.DataFrame({'boundary_score':boundary_s,'boundary_s_filtered':boundary_s_filtered,'strong_boundary':high_boundary,'smoothed_boundary_score':smoothed_score,'filmfest_offset':filmfest_offset,'TR_offset':tr_offset,'consensus_offset':consensus_offset})
    output_path='../../Analysis Data/behavior_measurements/event_boundary/'+'movie'+i+'_boundary_mark_mapped_to_neuro.csv'
    boundary_df.to_csv(output_path,index=False)

    plot_boundary(smd1,boundary_mark_filter)
  
```

## analysis

```python
def identify_value_around_boundary(df, selected_event, behavior):
    """
    Identifies behavior values immediately before and after event boundaries.
    
    Parameters:
        df (pd.DataFrame): Prediction data with 'offset' and behavior columns
        selected_event (pd.DataFrame): Event boundaries with 'consensus_offset'
        behavior (str): Column name of behavior to analyze
    
    Returns:
        tuple: (before_boundary, after_boundary, before_offset, after_offset)
               Arrays of behavior values and offsets surrounding boundaries
    """
    pred = df.sort_values("offset").reset_index(drop=True)
    pred_offsets = pred["offset"].values
    pred_consensus = pred[behavior].values
    target_offsets = selected_event["consensus_offset"].values
    
    n = len(selected_event)
    before_boundary = np.zeros(n)
    after_boundary = np.zeros(n)
    before_offset = np.zeros(n)
    after_offset = np.zeros(n)
    
    for i, val in enumerate(target_offsets):
        # Choose boundary side based on decimal: <=5 rounds down, >5 rounds up
        side = "right"
        if val % 10 <= 5:
            idx = np.searchsorted(pred_offsets, val, side=side) - 1
        else:
            idx = np.searchsorted(pred_offsets, val, side=side)
        
        idx = np.clip(idx, 0, len(pred_offsets) - 2)
        
        before_boundary[i] = pred_consensus[idx]
        after_boundary[i] = pred_consensus[idx + 1]
        before_offset[i] = pred_offsets[idx]
        after_offset[i] = pred_offsets[idx + 1]
    
    return before_boundary, after_boundary, before_offset, after_offset

```

```python
def filter_strong_boundaries_iterative(df, score_col='smoothed_boundary_score', 
                                      diff_threshold=30):
    """
    Filters nearby boundaries, keeping only the strongest in each cluster.
    
    Iteratively merges boundaries closer than diff_threshold, retaining
    the one with the highest score in each cluster.
    
    Parameters:
        df (pd.DataFrame): Boundary data with 'consensus_offset' and score_col
        score_col (str): Column to use for selecting strongest boundary
        diff_threshold (float): Maximum offset distance to merge boundaries
    
    Returns:
        pd.DataFrame: Filtered boundaries
    """
    df = df.sort_values('consensus_offset').copy().reset_index(drop=True)
    
    while True:
        df['diff'] = df['consensus_offset'].diff().fillna(diff_threshold + 1)
        
        # Cluster consecutive points closer than threshold
        clusters = []
        cluster = [0]
        for i in range(1, len(df)):
            if df['diff'].iloc[i] < diff_threshold:
                cluster.append(i)
            else:
                clusters.append(cluster)
                cluster = [i]
        clusters.append(cluster)
        
        # Keep highest-scoring point in each cluster
        keep_indices = [df.loc[cluster, score_col].idxmax() for cluster in clusters]
        new_df = df.loc[keep_indices].reset_index(drop=True)
        
        # Converged when no more merging occurs
        if len(new_df) == len(df):
            break
        df = new_df
    
    return df
```

```python
def plot_boundary_vs_behavior(df, selected_event, behavior,film):
    """
    Plots behavior over time with vertical lines at event boundaries.
    
    Parameters:
        df (pd.DataFrame): Data with 'offset' and behavior columns
        selected_event (pd.DataFrame): Event boundaries to mark
        behavior (str): Behavior column to plot
    """
    plt.figure(figsize=(10, 6))
    plt.plot(df['offset'], df[behavior], label=behavior, color='blue')
    
    for x in selected_event['consensus_offset']:
        plt.axvline(x=x, color='red', linestyle='--', alpha=0.7)
    
    plt.xlabel('Offset')
    plt.ylabel(behavior)
    plt.title(film.capitalize())
    plt.legend()
    plt.show()
```

```python
FILMS = ['theshoe', 'therock', 'theboyfriend', 'keithreynolds', 'cmiyc_long', 'busstop']
BOUNDARY_MAP = {
    'theshoe': '04', 'therock': '06', 'theboyfriend': '03',
    'keithreynolds': '05', 'cmiyc_long': '01', 'busstop': '10'
}

# Analysis
before_all = []
after_all = []
all_selected_events = []

for film in FILMS:
    # Load data
    prediction = pd.read_csv(f'../../Analysis Data/Prediction/{film}_consensus_mapped_to_neuro.csv')
    description = pd.read_csv(f'../../Analysis Data/Description/{film}_consensus_mapped_to_neuro.csv')
    boundary = pd.read_csv(f'../../Analysis Data/behavior_measurements/event_boundary/movie{BOUNDARY_MAP[film]}_boundary_mark_mapped_to_neuro.csv')
    
    # Exclude title and ending
    title, end = prediction['offset'].min(), prediction['offset'].max()
    prediction = prediction[(prediction['offset'] != title) & (prediction['offset'] != end)]
    description = description[(description['offset'] != title) & (description['offset'] != end)]
    #boundary=boundary[boundary['consensus_offset']>30]
    
    # Filter strong boundaries
    strong_boundary = boundary[boundary['boundary_score']>2].copy()
    selected_event = filter_strong_boundaries_iterative(
        strong_boundary, score_col='smoothed_boundary_score', diff_threshold=30
    )
    
    if selected_event.empty:
        continue
    
    # Z-score normalize accuracy
    #prediction['accuracy_z'] = zscore(prediction['accuracy'])
    
    # Extract boundary values
    before, after, before_off, after_off = identify_value_around_boundary(
        description, selected_event, "importance"
    )
    
    # Store results
    selected_event["before_boundary"] = before
    selected_event["after_boundary"] = after
    selected_event["before_offset"] = before_off
    selected_event["after_offset"] = after_off
    selected_event["film"] = film
    #visualize boundary locations
    plot_boundary_vs_behavior(description, selected_event, "importance",film)
    
    before_all.append(before)
    after_all.append(after)
    all_selected_events.append(selected_event)

# Aggregate results
before_all = np.concatenate(before_all)
after_all = np.concatenate(after_all)
final_selected_events = pd.concat(all_selected_events, ignore_index=True)
```

```python
# Statistical test
stat, p_value = wilcoxon(after_all - before_all, alternative='greater')
print(f"Wilcoxon test: stat = {stat}, p = {p_value}")
```
