## aggregate causality data

```python
def aggregate_causality_ratings(folder_path):
    """
    Aggregate causality ratings from multiple Excel files.
    
    Parameters:
    folder_path (str): Path to the folder containing the xlsx files
    
    Returns:
    pd.DataFrame: Aggregated dataframe with rating counts
    """
    # Find all xlsx files matching the pattern
    file_pattern = os.path.join(folder_path, 'filmfest_causality_*.xlsx')
    files = glob.glob(file_pattern)
    
    if not files:
        print(f"No files found matching pattern: {file_pattern}")
        return None
    
    print(f"Found {len(files)} files to process")
    
    # List to store dataframes from each file
    all_ratings = []
    
    # Read each file and extract the relevant columns
    for file in files:
        try:
            df = pd.read_excel(file)
            
            # Check if required columns exist
            required_cols = ['movie', 'cause_event', 'effect_event']
            if all(col in df.columns for col in required_cols):
                # Select only the required columns
                df_subset = df[required_cols].copy()
                all_ratings.append(df_subset)
                print(f"✓ Processed: {os.path.basename(file)}")
            else:
                print(f"✗ Skipped {os.path.basename(file)}: Missing required columns")
                
        except Exception as e:
            print(f"✗ Error reading {os.path.basename(file)}: {str(e)}")
    
    if not all_ratings:
        print("No valid data found in any files")
        return None
    
    # Combine all dataframes
    combined_df = pd.concat(all_ratings, ignore_index=True)
    
    # Group by the three columns and count occurrences
    aggregated_df = (combined_df.groupby(['movie', 'cause_event', 'effect_event'])
                     .size()
                     .reset_index(name='rating_count'))
    
    # Sort by rating_count in descending order
    aggregated_df = aggregated_df.sort_values('rating_count', ascending=False).reset_index(drop=True)
    
    print(f"\nAggregation complete!")
    print(f"Total unique causality pairs: {len(aggregated_df)}")
    print(f"Total ratings collected: {aggregated_df['rating_count'].sum()}")
    
    return aggregated_df

```

```python
# Specify your folder path here
folder_path = "../../Raw Data/causality ratings"

# Aggregate the ratings
result_df = aggregate_causality_ratings(folder_path)
if result_df is not None:
    # Display the first few rows
    result_df=result_df.sort_values(by=['movie','cause_event','effect_event'])
    print("\nFirst 10 rows of aggregated data:")
    print(result_df.head(10))
    
    # Optional: Save to CSV or Excel
    result_df.to_csv('../../Analysis Data/behavior_measurements/aggregated_causality_ratings.csv', index=False)
```

## analysis

```python
filename='../../Analysis Data/behavior_measurements/aggregated_causality_ratings.csv'
df=pd.read_csv(filename)
df_annotation = pd.read_csv('../../Raw Data/annotations/FilmFest_Annotation_JL_spellchecked_YJ.csv')
# keep only event that cause a later event
df=df[df['cause_event']<df['effect_event']]
```

```python
df_time=df_annotation[['SEG-216 concatenated', 'Start Time (m.ss)',
       'End Time (m.ss)']]
df_time=df_time.dropna()
df_time=df_time.rename(columns={'SEG-216 concatenated':'event_number'})
df_time['event_number']=df_time['event_number'].astype(int)
```

```python
# Convert time columns to seconds for easier calculation
def time_to_seconds(time_str):
    """Convert time in m.ss format to seconds"""
    if pd.isna(time_str):
        return None
    time_str = str(time_str)
    if '.' in time_str:
        minutes, seconds = time_str.split('.')
        if len(seconds)<2:
            return int(minutes) * 60 + int(seconds)*10

        return int(minutes) * 60 + int(seconds)
    else:
        return float(time_str) * 60  # Assume it's just minutes

# Apply conversion
df_time['start_seconds'] = df_time['Start Time (m.ss)'].apply(time_to_seconds)
df_time['end_seconds'] = df_time['End Time (m.ss)'].apply(time_to_seconds)

# Calculate duration
df_time['duration'] = df_time['end_seconds'] - df_time['start_seconds']
```

### 1. filter effect event that are not cause event, then see if prediction at begining of it will be better than later prediction

```python
df_merged = df.merge(df_time,how='left',right_on='event_number',left_on='effect_event')
n_ratings=1
# Get all unique cause events that are rated multiple times
result = df_merged.groupby(['cause_event'])['rating_count'].sum()
cause_events = result[result > n_ratings].index.tolist()
result = df_merged.groupby(['effect_event'])['rating_count'].sum()
effect_events = result[result > n_ratings].index.tolist()
movie_end = [26,41,67,80,106,136,153,164,180,216]

# Find effect events that:
# 1. Have duration > 10 seconds
# 2. Are not in the cause_event list

effect_events_filtered = df_merged[
    (df_merged['duration'] > 10) & 
    (~df_merged['effect_event'].isin(cause_events)) &
    (~df_merged['effect_event'].isin(movie_end)) &
    (df_merged['effect_event'].isin(effect_events))

]

# Display results
print(f"Number of effect events > 10s that are not cause events: {len(effect_events_filtered['effect_event'].unique())}")
print("\nUnique effect events meeting criteria:")
print(effect_events_filtered['effect_event'].unique())

# Optional: See the full filtered dataframe
effect_events_filtered[['movie', 'cause_event', 'effect_event', 'duration', 'Start Time (m.ss)', 'End Time (m.ss)']]
```

```python
# Function to find the closest cause event for each effect event
def keep_closest_cause(group):
    """For each effect event, keep only the row with the closest cause event"""        
    # Return the row with minimum distance
    return group.loc[group['cause_event'].idxmax()]

# Group by movie and effect_event, then apply the function
effect_events_closest = effect_events_filtered.groupby(['movie', 'effect_event']).apply(keep_closest_cause)

# Reset index to get a clean dataframe
effect_events_closest = effect_events_closest.reset_index(drop=True)
```

```python
def find_closest_accuracy_asof(effect_events_closest, combined_df):
    # Prepare the data
    effect_df = effect_events_closest.copy()
    effect_df['start_seconds']=effect_df['start_seconds'].astype(int)
    combined_sorted = combined_df.sort_values('filmfest_offset').reset_index(drop=True)
    
    # Create next row data
    combined_with_next = combined_sorted.copy()
    combined_with_next['next_accuracy'] = combined_sorted['accuracy'].shift(-1)
    
    # Merge asof to find closest matches
    result = pd.merge_asof(
        effect_df.sort_values('start_seconds'),
        combined_with_next[['filmfest_offset', 'accuracy', 'next_accuracy']],
        left_on='start_seconds',
        right_on='filmfest_offset',
        direction='nearest'
    )
    
    return result
```

```python
def find_closest_accuracy(effect_events_closest, combined_df,column):
    # Prepare the data
    effect_df = effect_events_closest.copy()
    effect_df[column]=effect_df[column].astype(int)
    combined_sorted = combined_df.sort_values('filmfest_offset').reset_index(drop=True)
        
    # Merge asof to find closest matches
    result = pd.merge_asof(
        effect_df.sort_values(column),
        combined_sorted[['filmfest_offset', 'accuracy']],
        left_on=column,
        right_on='filmfest_offset',
        direction='nearest'
    )
    
    return result
```

```python
result_df1 = find_closest_accuracy_asof(effect_events_closest[effect_events_closest['effect_event']<session1_end], combined_df1)

start_df1 = find_closest_accuracy(effect_events_closest[effect_events_closest['effect_event']<session1_end], combined_df1,'start_seconds')
end_df1 = find_closest_accuracy(effect_events_closest[effect_events_closest['effect_event']<session1_end], combined_df1,'end_seconds')
```

```python
def filter_rows(df,column):
    return df[abs(df['filmfest_offset'] - df[column]) < 10]
```

```python
result_df2 = find_closest_accuracy_asof(effect_events_closest[effect_events_closest['effect_event']>session1_end], combined_df2)
result_df2 = filter_rows(result_df2,'start_seconds')
start_df2 = find_closest_accuracy(effect_events_closest[effect_events_closest['effect_event']>session1_end], combined_df2,'start_seconds')
end_df2 = find_closest_accuracy(effect_events_closest[effect_events_closest['effect_event']>session1_end], combined_df2,'end_seconds')
start_df2 = filter_rows(start_df2,'start_seconds')
end_df2 = filter_rows(end_df2,'end_seconds')
```
